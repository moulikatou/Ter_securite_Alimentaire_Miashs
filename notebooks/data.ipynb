```python
import pyreadstat
import glob
import numpy as np
```

```python
#Téléchargement des fichiers SAV
sav_files = glob.glob('../Donnees_/*.sav')
sav_files
```

```python
years = [2014,2015,2018,2017,2016]
```

```python
#keep necessary cols
S62_Q1 = ["S62Q1_1","S62Q1_2","S62Q1_3","S62Q1_4","S62Q1_5"]
COL_TO_KEEP = ["REG", "COM", "PROV", "MEN","VIL", "YEAR"] #pas de 'MEN' in rcsi_2014.columns
COL_TO_KEEP += S62_Q1



#empty new df
newdf = pd.DataFrame(columns=COL_TO_KEEP)
```

```python
for année in range(len(sav_files)):
  
    nomdf, meta  = pyreadstat.read_sav(sav_files[année])
    nomdf['YEAR']= years[année]

    #print(len(nomdf.columns))
    
    if("CODMEN" in nomdf.columns):
        
        nomdf.rename(columns={"CODMEN" : "MEN"}, inplace = True)
        
    
    


    newdf = pd.concat([newdf, nomdf[COL_TO_KEEP ]])
```

```python
newdf = newdf.reset_index(drop=True)
newdf
```

```python
def rcsi_score(df_og):

    # recupération des réponses aux question en dataframe
    ans = df_og[S62_Q1]

    # pondération par leurs poids
    poids = [1,2,1,3,1]
    ans.loc[:, S62_Q1] *= np.array(poids)

    #(sum) rcsi par commune
    df_og['rcsi']=ans.sum(axis=1)
```

```python
rcsi_score(newdf)
newdf
```

### Computing rCsi moyen par province

```python
RcsiMoyProv = pd.DataFrame(columns=['Prov','Reg','year', 'Rcsi_moy'])
```

```python
years = [2014,2015,2018,2017,2016]
```

```python
listProv = result['Province'].unique()
```

```python
print(len(listProv))
listProv
```

```python
for i in range(len(listProv)):
    for y in range (len(years)):
        annee= years[y]
        df_annee = result[result['YEAR']== annee]
        prov_annee = df_annee[df_annee['Province']==listProv[i]]
        prov_annee.reset_index(drop=True, inplace=True)
        men = np.array(prov_annee['MEN'])
        rc = np.array(prov_annee['rcsi'])
        moy = np.average(rc, weights=men)
        
        print('nom région : ', prov_annee['Nom_region'][1])
        
        #moy = np.sum(RcsiMA *weight)
    
        new_row = {
            'Prov':listProv[i],
            'Reg':prov_annee['Nom_region'][1],
            'year':annee,
            'Rcsi_moy': moy   
        }

        RcsiMoyProv = RcsiMoyProv.append(new_row, ignore_index=True)
    
    
```
